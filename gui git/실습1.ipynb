```python
from tkinter import *
import tkinter as tk
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

root = tk.Tk()
rt1= tk.Tk()
rt2= tk.Tk()

def openrt2():
#rt2 디자인
    df1 = pd.read_csv('Adata.csv')
    figure2=plt.Figure(figsize=(5,4), dpi=100)
    ax2= figure2.add_subplot(111)
    line2=FigureCanvasTkAgg(figure2,rt2)
    line2.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    df2=df1['이용자수'].groupby(df1['읍면동명']).sum()
    df2.plot(kind='line',ax=ax2, color='r',marker='o',fontsize='10')
    ax2.set_title('sum of local')
    rt2.geometry('400x300+800+0')
    rt2.mainloop()
def openrt1():
    #rt1 디자인r=rt1)
    btn1=PhotoImage(file='bt3.png',master=rt1)
    btn2=PhotoImage(file='bt4.png',master=rt1)
    bg1=PhotoImage(file='apple_bg2.png',master=rt1)
    bg_label1=Label(rt1,image=bg1)
    bg_label1.place(x=0,y=0)
    rt1.geometry('300x500+500+0')
    rt1_bt1=Button(rt1,width=150, height = 50,image=btn1,command=closert2)#,command=closert1
    rt1_bt2=Button(rt1,width=150, height = 50,image=btn2,command=openrt2)#,command=openrt2
    rt1_bt1.place(x=70,y=300)
    rt1_bt2.place(x=70,y=400)

    rt1_lst=Listbox(rt1,selectmode='extended', height=0)
    rt1_lst.insert(0,'소매')
    rt1_lst.insert(0,'숙박')
    rt1_lst.insert(0,'식음료')
    rt1_lst.insert(0,'문화/레져')
    rt1_lst.place(x=75,y=200)
    rt1.mainloop()
    

def closert1():
    rt1.destroy()

def closert2():
    rt2.destroy()
    
    
root.geometry('500x750+0+0')
bg=PhotoImage(file='apple1.png',master=root)
bg_label=Label(root,image=bg)
bg_label.place(x=0,y=0)

bt_bg1=PhotoImage(file='bt1.png',master=root)
bt1=Button(root,width=150, height=50, image=bt_bg1,command=openrt1)
bt1.place(x=180, y=350)

bt_bg2=PhotoImage(file='bt2.png',master=root)
bt2=Button(root,width=150, height=50, image=bt_bg2)
bt2.place(x=180, y=450)

root.mainloop()
```
