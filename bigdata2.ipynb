```python
import pandas as pd
df1 = pd.read_csv('china01.csv')
print(df1.shape)
df1.head()
```

```python
df1
df1_data=df1.sort_values('업종')
df1_data
c=df1_data[df1_data['업종']=='교통'].index
c_=df1_data.drop(c)
ch=df1_data[df1_data['업종']=='기타'].index
c__=c_.drop(ch)
chi=df1_data[df1_data['업종']=='유흥'].index
c___=c__.drop(chi)
chin=df1_data[df1_data['업종']=='전체'].index
china_=c___.drop(chin)
china_
```

```python
china_=china_.drop(['년월','이용자 구분','시도명','업종','방문인구','이용금액','데이터기준일자',axis=1])
```

```python
data=pd.DataFrame(china_['이용자수'].groupby(china_['읍면동명']).sum())
data
```
