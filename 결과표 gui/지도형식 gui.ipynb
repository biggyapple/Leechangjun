```python
import tkinter as tk
from tkinter import *
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

df = pd.read_csv('Udong20.csv')
root=Tk()
root.geometry('1000x700+0+0') #root 창 크기  

bg=PhotoImage(file='png4.png',master=root)
bg_label=Label(root,image=bg)# 배경 레이블 생성
bg_label.place(x=0,y=0)

#두번째 레이블 생성
bg3=PhotoImage(file='png5.png',master=root)
label2=Label(root,image=bg3)
label2.pack(side=LEFT,padx=51)

#확인 버튼 생성
cbt=Button(root,width=75,height=3,text='확인')
cbt.pack(side=BOTTOM,pady=10)

figure=plt.Figure(figsize=(11,3), dpi=45) #Figure 객체 생성 및 그래프 크기
figure1=plt.Figure(figsize=(11,3), dpi=45) #Figure 객체 생성 및 그래프 크기
figure2=plt.Figure(figsize=(11,3), dpi=45) #Figure 객체 생성 및 그래프 크기
figure3=plt.Figure(figsize=(5.2,3), dpi=45) #Figure 객체 생성 및 그래프 크기
figure4=plt.Figure(figsize=(5.2,3), dpi=45) #Figure 객체 생성 및 그래프 크기

ax1= figure.add_subplot(1,1,1) #3-전체 행 , 1-그래프 보여줄 열 , 1-그래프를 보여줄 행 선택
ax2= figure1.add_subplot(1,1,1)
ax3= figure2.add_subplot(1,1,1)
ax4= figure3.add_subplot(1,1,1)
ax5= figure4.add_subplot(1,1,1)

line1=FigureCanvasTkAgg(figure,root) 
line1.get_tk_widget().pack(side=BOTTOM,pady=5)
line2=FigureCanvasTkAgg(figure1,root) 
line2.get_tk_widget().pack(side=BOTTOM,pady=5)
line3=FigureCanvasTkAgg(figure2,root) 
line3.get_tk_widget().pack(side=BOTTOM,pady=5)
line4=FigureCanvasTkAgg(figure3,root) 
line4.get_tk_widget().place(x=479,y=50)
line5=FigureCanvasTkAgg(figure4,root) 
line5.get_tk_widget().place(x=741,y=50)

#첫번째 레이블 생성
label1=Label(root,width=76,height=2,text='소매업에 추천드릴 지역은 oo 입니다. ')
label1.place(x=459,y=5)

df1=df['이용자수'].groupby(df['읍면동명']).sum() #보여줄 그래프
df1.plot(kind='line',ax=ax1, color='g',marker='o',fontsize='10') # 선 그래프 형태
ax1.set_title('Line graph')#그래프 제목

df2=df['이용자수'].groupby(df['읍면동명']).sum()
df2.plot(kind='bar',ax=ax2, color='r',fontsize='10') # 막대 그래프 형태
ax2.set_title('Bar graph')

df3=df['이용자수'].groupby(df['읍면동명']).sum()
df3.plot(kind='hist',ax=ax3, color='b',fontsize='10')
ax3.set_title('sum of local')

df4=df['이용자수'].groupby(df['읍면동명']).sum()
df4.plot(kind='hist',ax=ax4, color='g',fontsize='10')
ax4.set_title('sum of local')

df5=df['이용자수'].groupby(df['읍면동명']).sum()
df5.plot(kind='bar',ax=ax5, color='r',fontsize='10')
ax5.set_title('sum of local')
plt.show()


root.mainloop()
```
