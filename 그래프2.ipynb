```python
import tkinter as tk
from tkinter import *
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

df = pd.read_csv('Udong20.csv')
root=Tk()
root.geometry('500x800+0+0') #root 창 크기  

bg=PhotoImage(file='png1.png',master=root)
bg_label=Label(root,image=bg)
bg_label.place(x=0,y=0)

label=Label(root,width=50,height=30,text='소매업 관련 정보')
label.place(x=0,y=0)

figure=plt.Figure(figsize=(6,3), dpi=100) #Figure 객체 생성 및 그래프 크기
ax1= figure.add_subplot(3,1,1) #3-전체 행 , 1-그래프 보여줄 열 , 1-그래프를 보여줄 행 선택
ax2= figure.add_subplot(3,1,2)
ax3= figure.add_subplot(3,1,3)
line1=FigureCanvasTkAgg(figure,root) 
line1.get_tk_widget().pack(side=tk.LEFT,fill=X)
line2=FigureCanvasTkAgg(figure,root)
line2.get_tk_widget().pack(side=tk.LEFT,fill=X)
line3=FigureCanvasTkAgg(figure,root)
line3.get_tk_widget().pack(side=tk.LEFT,fill=X)

df1=df['이용자수'].groupby(df['읍면동명']).sum() #보여줄 그래프
df1.plot(kind='line',ax=ax1, color='r',marker='o',fontsize='10') # 선 그래프 형태
ax1.set_title('Line graph')#그래프 제목

df2=df['이용자수'].groupby(df['읍면동명']).sum()
df2.plot(kind='bar',ax=ax2, color='r',fontsize='10') # 막대 그래프 형태
ax2.set_title('Bar graph')

df3=df['이용자수'].groupby(df['읍면동명']).sum()
df3.plot(kind='hist',ax=ax3, color='r',fontsize='10')
ax3.set_title('sum of local')
plt.show()


root.mainloop()
```
